#### False means no missing data
- df.isnull().sum()int
- df.isnull().any()bool

```python
# To determine which values are null and not null
df.isnull()
```

```python
# To call out only the null values
df.isnull().any()
```

```python
# Columns with NaN using True/False
# False means it doesn't have a NaN
df.isnull().any().any()
```

```python
df.head()
```

```python
# Filling it with mean
df['DURATION'].mean()
```
